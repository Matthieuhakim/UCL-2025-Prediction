```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
import time

standings_url = "https://fbref.com/en/comps/8/Champions-League-Stats"
```

```python
data = requests.get(standings_url)
soup = BeautifulSoup(data.text, "html.parser")
```

```python
standings_table = soup.select('table.stats_table')[0]
links = standings_table.find_all('a')
team_urls = [f"https://fbref.com{l.get('href')}" for l in links if l.get('href') and '/squads/' in l.get('href')]
all_teams_data = []
```

```python
for team_url in team_urls:
  print(f"Processing {team_url}")
  time.sleep(2)

  data = requests.get(team_url)
  soup = BeautifulSoup(data.text, "html.parser")

  # Get matches data
  try:
      matches = pd.read_html(data.text, match='Scores & Fixtures')[0]
      matches = matches[matches["Comp"] == "Champions Lg"]  # Filter only Champions League matches
  except ValueError:
      print(f"No match data found for {team_url}")
      continue

  # Get shooting data
  links = soup.find_all('a')
  shooting_links = [f"https://fbref.com{l.get('href')}" for l in links if l.get('href') and 'c8/shooting/' in l.get('href')]

  time.sleep(2)
  if shooting_links:
      shooting_url = shooting_links[0]
      data = requests.get(shooting_url)
      try:
          shooting = pd.read_html(data.text, match='Shooting')[0]
          shooting.columns = shooting.columns.droplevel()
      except ValueError:
          print(f"No shooting data found for {team_url}")
          shooting = pd.DataFrame()
  else:
      shooting = pd.DataFrame()

  # Merge match and shooting data
  if not shooting.empty:
      team_data = matches.merge(shooting[["Date", "Sh", "SoT", "Dist", "FK", "PK", "PKatt"]], on="Date", how="left")
  else:
      team_data = matches

  team_data["Team URL"] = team_url  # Add team reference
  all_teams_data.append(team_data)
```

```python
# Combine all teams' data into one DataFrame
final_df = pd.concat(all_teams_data, ignore_index=True)

# Save to CSV
final_df.to_csv("champions_league_stats.csv", index=False)
print("Data saved to champions_league_stats.csv")
```
