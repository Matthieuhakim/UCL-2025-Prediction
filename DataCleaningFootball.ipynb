```python
import numpy as np
import pandas as pd
```

```python
matches = pd.read_csv("champions_league_stats.csv")
matches.head()
```

```python
matches.shape
```

```python
matches.columns
```

```python
matches = matches[matches["Round"] == "League phase"]
matches.shape
```

```python
matches = matches[['Team URL', 'Opponent', 'Date', 'Venue', 'Result', 'GF', 'GA', 'xG', 'xGA', 'Poss', 'Sh', 'SoT', 'Dist', 'FK', 'PK', 'PKatt']]
matches.head()
```

```python
# Assign Unique ID for teams

team_urls = matches['Team URL'].unique()
team_ids = range(len(team_urls))
team_map = dict(zip(team_urls, team_ids))

print(team_map)

# Replace team URLs with team IDs
matches['Team ID'] = matches['Team URL'].map(team_map)

# Drop the original 'Team URL' column
matches = matches.drop('Team URL', axis=1)

matches.head()
matches.shape
matches.columns
```

```python
# Create 'Opponent ID' column and initialize it to NaN
matches['Opponent ID'] = np.nan

# Match Opponent to
for index, row in matches.iterrows():
    opponent = row['Opponent']
    opponent = opponent.split()
    opponent = opponent[1:]
    opponent = "-".join(opponent)

    # Fix Exceptions
    if "Paris" in opponent:
        opponent = "Paris-Saint-Germain"
    elif "Atlético" in opponent:
        opponent = "Atl"
    elif "Real" in opponent:
        opponent = "Real"
    elif "RB-Salzburg" in opponent:
        opponent = "Salzburg"

    for team_url, team_id in team_map.items():
        if opponent in team_url:
            matches.loc[index, 'Opponent ID'] = int(team_id)
            break  # Exit inner loop once a match is found
    else:
        print(f"No match found for opponent: {opponent}")

# Drop the original 'Opponent' column
matches = matches.drop('Opponent', axis=1)
```

```python
matches.isnull().sum()
```

```python
matches.head()
```

```python
# Convert result to points
matches['Points'] = matches['Result'].map({'W': 3, 'D': 1, 'L': 0})

# Get Goal Difference
matches['Goal Difference'] = matches['GF'] - matches['GA']

# Get Expected Goal Difference
matches['xG Difference'] = matches['xG'] - matches['xGA']

# Get Shot Accuracy
matches['Shot Accuracy'] = matches['SoT'] / matches['Sh']

```

```python
matches['Date'] = pd.to_datetime(matches['Date'])

matches.head()
```

```python
matches.to_csv('ucl_match_data.csv', index=False)
```
